```python
import pandas as pd
```

```python
df = pd.read_csv('aula_0_08.csv', encoding = 'UTF-8', sep = ';')
```

```python
df.head(4)
```

```python
# Retorna as pessoas solteiras do df
df[df['marital'] == 'single'].head(4)
```

```python
# Retorna as pessoas solteiras E com educação primária
df[(df['marital'] == 'single') & (df['education'] == 'primary')].head(4)
```

```python
# Retorna as pessoas solteiras OU com educação secundária
df[(df['marital'] == 'single') | (df['education'] == 'secondary')].head(4)
```
