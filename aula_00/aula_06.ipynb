```python
#Ordenar dados usando sort_valoues e parametros como inplace (armazena a alteração) e ascending (ordena de forma crescente ou descrescente)
```

```python
import pandas as pd
```

```python
df = pd.read_excel('aula_06.xlsx')
```

```python
df.head(4)
```

```python
# Ordena os valores conforme coluna FirstName
df.sort_values('FirstName')
```

```python
# Ordena valores conforme coluna FirstName e salva as alterações (inplace)
df.sort_values('FirstName', inplace = True)
```

```python
df.head(4)
```

```python
# Ordena valores conforme colunas FirstName e LastName e ordena crescente e decrescente respectivamente
df.sort_values(by = ['FirstName', 'LastName'], ascending = [True, False]).head()
```
