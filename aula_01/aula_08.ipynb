```python
import pandas as pd
import numpy as np
```

```python
import os
from pathlib import Path
p = Path(os.getcwd())
df = pd.read_csv(str(p.parent) + '\\bases_de_dados\\houses_to_rent.csv')
```

```python
df.head(3)
```

```python
# Renomear colunas
colunas_anteriores = df.columns.to_list()
colunas_novas = ['indice', 'cidade', 'area_m2', 'quartos', 'banheiros', 'vagas_estacionamento', 
                 'andar', 'aceita_animais', 'mobiliado', 'valor_aluguel', 'taxas', 
                 'seguro_incendio', 'total']
colunas = dict((chave, valor) for chave, valor in zip(colunas_anteriores, colunas_novas))

df.rename(columns = colunas, inplace = True)
```
